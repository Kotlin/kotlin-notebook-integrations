```kotlin
val itemsHBox = List(4) { widgetManager.label { value = it.toString() } }
widgetManager.hBox {
    children = itemsHBox
}
```

```kotlin
val items = List(4) { widgetManager.label { value = it.toString()} }
val leftBox = widgetManager.vBox { children = listOf(items[0], items[1]) }
val rightBox = widgetManager.vBox { children = listOf(items[2], items[3]) }
widgetManager.hBox { children = listOf(leftBox, rightBox) }
```

```kotlin
val accordion = widgetManager.accordion {
    children = listOf(widgetManager.intSlider(), widgetManager.text())
    titles = listOf("Slider", "Text")
}
accordion
```

```kotlin
val tabContents = listOf("P0", "P1", "P2", "P3", "P4")
val _children = tabContents.map { widgetManager.text { description = it } }
val tab = widgetManager.tab {
    this.children = _children
    titles = _children.indices.map { it.toString() }
}
tab
```

```kotlin
val tabNest = widgetManager.tab {
    children = listOf(accordion, accordion)
    titles = listOf("An accordion", "Copy of the accordion")
}
tabNest
```
